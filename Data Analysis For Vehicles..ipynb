```python
# Using the interquartile range
 # 75th percentile
seventy_fifth = data["year"].quantile(0.75)
# 25th percentile
twenty_fifth = data["year"].quantile(0.25)
# Interquartile range
data_iqr = seventy_fifth - twenty_fifth
print(data_iqr)
# Identifying outliers
# Upper threshold
upper = seventy_fifth + (1.5 * data_iqr)
# Lower threshold
lower = twenty_fifth - (1.5 * data_iqr)
print(upper, lower)
# Subsetting our data
print(data[(data["year"] < lower) | (data["year"] > upper)] \
[["year","fuel","model"]])
no_outliers = data[(data["year"] > lower) & (data["year"] < upper)]

print(no_outliers["year"].describe())
sns.boxplot(data=no_outliers,x="year")
plt.show()

```

```python
 # Price Analysis:
   #- What is the average price of vehicles in the dataset? How does it vary across different regions and manufacturers?
   #- Are there any notable trends or patterns in vehicle prices over the years?
   #- How does the price differ based on the vehicle's fuel type, transmission type, or title status?
```

```python
# Calculate the average price of vehicles in the dataset
average_price = data['price'].mean()
print("Average price of vehicles:", average_price)

# Graph: Histogram of Price Distribution
sns.histplot(data=data, x='price', bins=20)
plt.xlabel('Price')
plt.ylabel('Count')
plt.title('Distribution of Vehicle Prices')
plt.show()
```

```python
# How does it vary across different regions and manufacturers?
# Calculate average price by region and manufacturer
average_price_by_region = data.groupby('region')['price'].mean()
average_price_by_manufacturer = data.groupby('manufacturer')['price'].mean()

# Graph: Bar Plot of Average Price by Region
plt.figure(figsize=(12, 6))
sns.barplot(data=average_price_by_region.reset_index(), x='region', y='price')
plt.xlabel('Region')
plt.ylabel('Average Price')
plt.title('Average Vehicle Price by Region')
plt.xticks(rotation=90)
plt.show()

# Graph: Bar Plot of Average Price by Manufacturer
plt.figure(figsize=(12, 6))
sns.barplot(data=average_price_by_manufacturer.reset_index(), x='manufacturer', y='price')
plt.xlabel('Manufacturer')
plt.ylabel('Average Price')
plt.title('Average Vehicle Price by Manufacturer')
plt.xticks(rotation=90)
plt.show()
```

```python
# Calculate average price by year
average_price_by_year = data.groupby('year')['price'].mean()

# Graph: Line Plot of Average Price over Years
plt.figure(figsize=(12, 6))
sns.lineplot(data=average_price_by_year.reset_index(), x='year', y='price')
plt.xlabel('Year')
plt.ylabel('Average Price')
plt.title('Average Vehicle Price over Years')
plt.xticks(rotation=45)
plt.show()
```

```python
# Calculate average price by fuel type, transmission type, and title status
average_price_by_fuel = data.groupby('fuel')['price'].mean()
average_price_by_transmission = data.groupby('transmission')['price'].mean()
average_price_by_title_status = data.groupby('title_status')['price'].mean()

# Graph: Bar Plot of Average Price by Fuel Type
plt.figure(figsize=(12, 6))
sns.barplot(data=average_price_by_fuel.reset_index(), x='fuel', y='price')
plt.xlabel('Fuel Type')
plt.ylabel('Average Price')
plt.title('Average Vehicle Price by Fuel Type')
plt.xticks(rotation=45)
plt.show()

# Graph: Bar Plot of Average Price by Transmission Type
plt.figure(figsize=(12, 6))
sns.barplot(data=average_price_by_transmission.reset_index(), x='transmission', y='price')
plt.xlabel('Transmission Type')
plt.ylabel('Average Price')
plt.title('Average Vehicle Price by Transmission Type')
plt.xticks(rotation=45)
plt.show()

# Graph: Bar Plot of Average Price by Title Status
plt.figure(figsize=(12, 6))
sns.barplot(data=average_price_by_title_status.reset_index(), x='title_status', y='price')
plt.xlabel('Title Status')
plt.ylabel('Average Price')
plt.title('Average Vehicle Price by Title Status')
plt.xticks(rotation=45)
plt.show()
```

```python
# Calculate the average and median odometer readings
average_odometer = data['odometer'].mean()
median_odometer = data['odometer'].median()
print("Average Odometer Reading:", average_odometer)
print("Median Odometer Reading:", median_odometer)

# Graph: Histogram of Odometer Readings
sns.histplot(data=data, x='odometer', bins=20)
plt.xlabel('Odometer Reading')
plt.ylabel('Count')
plt.title('Distribution of Odometer Readings')
plt.show()
```

```python
# Calculate the average odometer readings by vehicle manufacturer
average_odometer_by_manufacturer = data.groupby('manufacturer')['odometer'].mean()

# Graph: Bar Plot of Average Odometer Readings by Manufacturer
plt.figure(figsize=(12, 6))
sns.barplot(data=average_odometer_by_manufacturer.reset_index(), x='manufacturer', y='odometer')
plt.xlabel('Manufacturer')
plt.ylabel('Average Odometer Reading')
plt.title('Average Odometer Reading by Manufacturer')
plt.xticks(rotation=90)
plt.show()

# Calculate the average odometer readings by vehicle model
average_odometer_by_model = data.groupby('model')['odometer'].mean().head(20)

# Graph: Bar Plot of Average Odometer Readings by Model
plt.figure(figsize=(12, 6))
sns.barplot(data=average_odometer_by_model.reset_index(), x='model', y='odometer')
plt.xlabel('Model')
plt.ylabel('Average Odometer Reading')
plt.title('Average Odometer Reading by Model')
plt.xticks(rotation=90)
plt.show()
```

```python
# Calculate the average odometer readings by vehicle year
average_odometer_by_year = data.groupby('year')['odometer'].mean()

# Graph: Line Plot of Average Odometer Readings by Year
plt.figure(figsize=(12, 6))
sns.lineplot(data=average_odometer_by_year.reset_index(), x='year', y='odometer')
plt.xlabel('Year')
plt.ylabel('Average Odometer Reading')
plt.title('Average Odometer Reading by Year')
plt.xticks(rotation=45)
plt.show()

# Graph: Scatter Plot of Odometer Reading vs. Vehicle Year
plt.figure(figsize=(10, 6))
sns.scatterplot(data=data, x='year', y='odometer')
plt.xlabel('Year')
plt.ylabel('Odometer Reading')
plt.title('Odometer Reading vs. Vehicle Year')
plt.show()
```
